# Analise Descritiva

## Bibliotecas

```python
import pandas as pd 
import numpy as np 
import scipy 
import matplotlib.pyplot as plt 
import seaborn as sns
```

## Importando dados 

```python
df = pd.read_csv('pov19ce_clean.csv')
```

## Analise descritiva

### Rotulos por variaveis

```python
features = {}
for feature in df.columns:
    if str(df[feature].dtype) != 'float64' and str(df[feature].dtype) != 'int64':
        features[feature]= list(df[feature].unique())
features
```

### Histogramas com graficos 

```python
df.hist(figsize=(20, 20));
```

### Indicador de Proporo de Pobres 

```python
ax = df['pobre'].value_counts().sort_values().plot(kind="barh")
totals= []
for i in ax.patches:
    totals.append(i.get_width())
total = sum(totals)
for i in ax.patches:
     ax.text(i.get_width()+.3, i.get_y()+.20, 
     str(round((i.get_width()/total)*100, 1))+'%', 
     fontsize=10, color='black')
ax.grid(axis="x")
ax.set_xlim([0, 6000])
plt.title("Pobres", fontsize=20)
plt.show()
```
